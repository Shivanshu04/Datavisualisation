```python
# Relation Between 'sepal_length'and 'sepal_width'
# here we plot the scatter plot which means dot plot
# from the graph we can say that data set is completely distributed and there is no relation
df.plot.scatter(x ='sepal_length' , y ='sepal_width'  )
```

```python
#Relation between 'sepal_length' and 'petal_length'
# Here from this graph we can say that when sepal_length is increasing petal_length is also increasing.
df.plot.scatter(x ='sepal_length' , y ='petal_length'  )
```

```python
#Relation between 'sepal_length' and 'sepal_width'
df.plot.scatter(x ='sepal_length' ,y='petal_length', c ='sepal_width'  )
# This 3rd parameter decides the colour of the dot in this particular graph.
```

```python
df
```

```python
# The 4th parameter 's = 100' means size of the dot we can increase and decrease the sixe accordingly.
df.plot.scatter(x ='sepal_length' , y ='petal_length' , c =  'sepal_width', s = 100)
```

```python
#How to make size of the dot(which is s) as a dynamic parameter.
# Here the size 's' varies with the 'sepal length' we can assign a constant value or a variable.
df.plot.scatter(x ='sepal_length' , y ='petal_length' , c =  'sepal_width', s = df['sepal_length']*20)
```
