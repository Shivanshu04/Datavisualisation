```python
# Relation Between 'sepal_length'and 'sepal_width'
# here we plot the scatter plot which means dot plot
# from the graph we can say that data set is completely distributed and there is no relation
df.plot.scatter(x ='sepal_length' , y ='sepal_width'  )
```

```python
#Relation between 'sepal_length' and 'petal_length'
# Here from this graph we can say that when sepal_length is increasing petal_length is also increasing.
df.plot.scatter(x ='sepal_length' , y ='petal_length'  )
```

```python
df
```

```python
# The 4th parameter 's = 100' means size of the dot we can increase and decrease the sixe accordingly.
df.plot.scatter(x ='sepal_length' , y ='petal_length' , c =  'sepal_width', s = 100)
```

```python
#How to make size of the dot(which is s) as a dynamic parameter.
# Here the size 's' varies with the 'sepal length' we can assign a constant value or a variable.
df.plot.scatter(x ='sepal_length' , y ='petal_length' , c =  'sepal_width', s = df['sepal_length']*20)
```

```python
# In this case all the dots are represented in the form of hexabin. and we can control the grid size using 'gridsize' parameter.
df.plot.hexbin(x ='sepal_length' , y ='petal_length',gridsize=10)
```

```python
# cmap="viridis" this means colour map which we use to choose any colour of our choice. which is optional. we use it to create a fancy graph.
df.plot.hexbin(x ='sepal_length' , y ='petal_length',gridsize=10,cmap="viridis")
```

3-D Plot
# with the help of mplot3d we will be able to plot 3d graphs
from mpl_toolkits import mplot3d

create data points for x-axis and y-axis and z-axis since it is 3d graph.

```python
# we use linespace to generate evenly spaced numbers over a specified interval.
# It will generate the 30 evenly spaced data points b/w 1 to 6. 
x = np.linspace(-1,6 , 30)
x
```

```python
y = np.linspace(-1,6 , 30 )
y
```

```python
z=x+y
z
```

```python
#plot a 3d graph
ax = plt.axes(projection = '3d') # here we are trying to create the axis(i.e the 3d enviornment(white colour grid enviornment) we are trying to create with the help of this)
ax.plot3D(x,y,z) # Here we are giving x,y and z coordinate. First it creates the dots then it creates line there.
ax.set_xlabel('x') # Here we just label the x axis lly we set y and z axis. so that it will show which one is x-axis and so on.
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
df
```

```python
#Create a 3d graph for sepal_length, sepal_width and petal_length in iris dataset.
ax = plt.axes(projection = '3d')
ax.plot3D(df['sepal_length'],df['sepal_width'],df['petal_length'])
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('petal_length')
```

Histogram Graph

```python
# Histogram which represents range on x-axis and frequency on y-axis i.e no. of data points in particular range..
# In the below graph how many data points are present between range 0 to 1? --> Ans 50 approx
# It is similar to value_count or group_by count operation
df.plot(kind = 'hist',figsize = (20,10))
```

```python
df
```

```python
#Plot histogram for a particular column
# From the graph we can concluse that actual data inside 'sepal_length' lie b/w 3 to 7.8 approx.
# What is the range where majority of data points lie?--> approx 5.4 to 5.7
df['sepal_length'].plot(kind = 'hist', figsize = (20,10))
```

```python
# when we call hist() function over a data frme it gives the histogram of all the numerical data.where 'alpha' is the density of color which lies between 0 to 1.
df.hist(figsize = (20,10), color = 'b', alpha = 0.8)
```

```python
# To choose a specific color go to this website 'https://uicolorpicker.com/' and choose one.
df.hist(figsize = (20,10), color = '#E6425E', alpha = 0.8)
```
